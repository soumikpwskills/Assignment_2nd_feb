**Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.**

**Ans :-**

####DataFrame :-
Pandas DataFrame is two-dimensional size-mutable, potentially heterogeneous tabular data structure with labeled axes (rows and columns). A Data frame is a two-dimensional data structure, i.e., data is aligned in a tabular fashion in rows and columns. Pandas DataFrame consists of three principal components, the data, rows, and columns.

A Python one-dimensional labelled array called a Pandas Series may hold any form of data (integer, float, string, etc.). It resembles a table in a database or a column in a spreadsheet. Each component of a series has a unique identification thanks to an index. It is possible to create new Series by using lists, arrays, dictionaries, and existing Series objects. They are an essential component of the Pandas library and are commonly used for data manipulation and analysis tasks. The more complex Pandas DataFrame data structure, which resembles a two-dimensional table and is composed of multiple Series objects, also heavily relies on Series

```python
#pandas.series

import pandas as pd

data = [1000, 2000, 3000, 4000, 5000]
s = pd.Series(data)
s
```

```python
#DataFrame
import pandas as pd

data = [1000, 2000, 3000, 4000, 5000]
df = pd.DataFrame(data, columns=['Column1'])
df
```

**Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can you give an example of when you might use one of these functions?**

**Ans :-**
1. pd.read_csv, pd.read_excel :-
The first function to mention is read_csv or read_excel. The functions are self-explanatory already. They are used to read a CSV or an excel file to a pandas DataFrame format.

`df = pd.read_csv("sample.csv")`

`df1= pd.red_excel("sample.xlsx")`

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

2. df.head() :- It is also common to use .head() function after read_csv or read_excel to see the data frame. By default, it shows the first 5 rows of the DataFrame.

```python
df.head()
```

3. df.dtypes :-
This is another very basic and widely used functions. Because it is necessary to know the data types of the variables before we dive into the analysis, visualization, or predictive modeling.

```python
df.dtypes
```

4. df.columns :-
When you have a big dataset like that it can be hard to see all the columns. Using .columns function, we can print out all the columns of the dataset.

```python
df.columns
```

5. df.describe() :-
This is a great function that provides some basic statistical measures such as count, mean, standard deviation etc.

```python
df.describe()
```

6. df.iloc():-
This function takes as a parameter the rows and column indices and gives you the subset of the DataFrame accordingly.

```python
df.iloc[0:10]
```

7. df.loc() :-
This function does almost the similar operation as .iloc() function. But here we can specify exactly which row index we want and also the name of the columns we want in our subset.

```python
df.loc[0:10]
```

**Q7. Create a DataFrame using multiple Series. Explain with an example.**

```python
import pandas as pd

student = ['Jitender', 'Purnima','Arpit', 'Jyoti']
rollno = [10, 11, 14, 78]

student_series = pd.Series(student)
rollno_series = pd.Series(rollno)

frame = {'Student': student_series,
		'Roll No.': rollno_series}

result = pd.DataFrame(frame)

print(result)
```
